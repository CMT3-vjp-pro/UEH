## 1.1. THỐNG KÊ MÔ TẢ

### 1.1.1. Ôn tập lý thuyết

1. **Thống kê mô tả là gì? Khác gì với thống kê suy luận?**
   - *Thống kê mô tả* tóm tắt và mô tả đặc điểm của tập dữ liệu hiện có (mean, median, std...).
   - *Thống kê suy luận* dùng mẫu để suy luận về tổng thể (kiểm định giả thuyết, khoảng tin cậy...).

2. **Khi nào dùng trung vị thay vì trung bình?**
   - Khi dữ liệu có giá trị ngoại lai (outliers) hoặc phân bố lệch, trung vị bền vững hơn.

3. **Các loại phân bố phổ biến:** phân bố chuẩn (normal), lệch trái (left-skewed), lệch phải (right-skewed).

4. **Q1, Q2, Q3 trong boxplot:** Q1=25th percentile, Q2=median=50th, Q3=75th percentile. IQR = Q3 - Q1.

5. **Xử lý outliers:** dùng IQR rule (< Q1 - 1.5×IQR hoặc > Q3 + 1.5×IQR) hoặc Z-score để phát hiện và xử lý.

### 1.1.3. Bài tập thực hành 1 - Thống kê mô tả: Red Wine Quality
> Dataset: https://www.kaggle.com/code/eisgandar/red-wine-quality-eda-classification  

```python
import numpy as np
import pandas as pd
from scipy import stats
# Load dữ liệu rượu đỏ
wine_data = pd.read_csv("data/winequality-red.csv")
print(wine_data.head())
print(wine_data.shape)
print(wine_data.dtypes)
```

```python
# Thống kê mô tả tổng quan
print(wine_data.describe())
```

```python
# Tính chi tiết cho cột 'alcohol'
col = wine_data['alcohol']
print(f"Mean: {np.mean(col):.4f}")
print(f"Median: {np.median(col):.4f}")
print(f"Mode: {stats.mode(col).mode:.4f}")
print(f"Variance: {np.var(col):.4f}")
print(f"Std Dev: {np.std(col):.4f}")
print(f"Min: {np.min(col):.4f}")
print(f"Max: {np.max(col):.4f}")
print(f"Range: {np.max(col) - np.min(col):.4f}")
print(f"Q1 (25th): {np.percentile(col, 25):.4f}")
print(f"Q2 (50th): {np.percentile(col, 50):.4f}")
print(f"Q3 (75th): {np.percentile(col, 75):.4f}")
print(f"IQR: {stats.iqr(col):.4f}")
# Kiểm tra giá trị thiếu
print(wine_data.isnull().sum())
```

### 1.2.1. Ôn tập lý thuyết

- **Histogram**: phân phối dữ liệu số liên tục
- **Bar chart**: so sánh dữ liệu phân loại
- **Boxplot**: phân tán, tứ phân vị, outliers
- **Scatter plot**: mối quan hệ giữa 2 biến số
- **Matplotlib**: low-level, linh hoạt cao
- **Seaborn**: high-level, tích hợp với pandas, đẹp hơn mặc định
- **Plotly**: interactive charts

### 1.2.3. Bài tập thực hành 1 - Trực quan hóa: Red Wine Quality

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
wine_data = pd.read_csv("data/winequality-red.csv")
print(wine_data.head())
print(wine_data.shape)
```

```python
# Histogram - phân phối của 'alcohol'
plt.figure(figsize=(10, 5))
ax = sns.histplot(data=wine_data, x='alcohol', kde=True)
ax.set_xlabel('Alcohol Content (%)', fontsize=13)
ax.set_ylabel('Count', fontsize=13)
ax.set_title('Distribution of Alcohol Content in Red Wine', fontsize=15)
plt.tight_layout()
plt.show()
```

```python
# Boxplot - phân bố các đặc trưng chính
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
cols = ['fixed acidity', 'volatile acidity', 'alcohol', 'pH', 'density', 'quality']
for i, col in enumerate(cols):
    r, c = divmod(i, 3)
    sns.boxplot(data=wine_data, y=col, ax=axes[r][c])
    axes[r][c].set_title(f'Boxplot of {col}', fontsize=11)
plt.tight_layout()
plt.show()
```

```python
# Bar chart - số lượng theo quality
plt.figure(figsize=(9, 5))
ax = sns.countplot(data=wine_data, x='quality', palette='viridis')
ax.set_xlabel('Wine Quality Score', fontsize=13)
ax.set_ylabel('Count', fontsize=13)
ax.set_title('Distribution of Wine Quality', fontsize=15)
plt.tight_layout()
plt.show()
```

#### Trực quan Online Retail

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

retail = pd.read_excel("data/Online Retail.xlsx", sheet_name='Online Retail')
print(retail.shape)
print(retail.dtypes)
print(retail.head())

# Histogram tất cả cột số
retail.select_dtypes(include='number').hist(figsize=(12, 6), bins=30, color='steelblue', edgecolor='white')
plt.suptitle('Histogram of Numeric Features - Online Retail', fontsize=15, y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Tiền xử lý: loại bỏ returns (Quantity < 0), UnitPrice <= 0, CustomerID null
retail_clean = retail.dropna(subset=['CustomerID'])
retail_clean = retail_clean[(retail_clean['Quantity'] > 0) & (retail_clean['UnitPrice'] > 0)]
retail_clean = retail_clean.copy()

# Tính doanh thu từng dòng
retail_clean['Revenue'] = retail_clean['Quantity'] * retail_clean['UnitPrice']

# Chuyển InvoiceDate sang datetime và trích xuất tháng
retail_clean['InvoiceDate'] = pd.to_datetime(retail_clean['InvoiceDate'])
retail_clean['YearMonth'] = retail_clean['InvoiceDate'].dt.to_period('M')

print(f'Sau lọc: {retail_clean.shape[0]:,} dòng')
print(retail_clean[['Quantity', 'UnitPrice', 'Revenue']].describe().round(2))
```

```python
# Top 10 quốc gia theo doanh thu
country_revenue = (
    retail_clean.groupby('Country')['Revenue']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

plt.figure(figsize=(12, 6))
ax = sns.barplot(data=country_revenue, x='Revenue', y='Country', palette='Blues_r')
ax.set_title('Top 10 Countries by Revenue - Online Retail', fontsize=15)
ax.set_xlabel('Tổng doanh thu (GBP)', fontsize=12)
ax.set_ylabel('Quốc gia', fontsize=12)
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))
plt.tight_layout()
plt.show()
```

```python
# Top 10 sản phẩm bán chạy nhất theo số lượng
top_products = (
    retail_clean.groupby('Description')['Quantity']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

plt.figure(figsize=(12, 6))
ax = sns.barplot(data=top_products, x='Quantity', y='Description', palette='Greens_r')
ax.set_title('Top 10 Best-Selling Products by Quantity', fontsize=15)
ax.set_xlabel('Tổng số lượng bán ra', fontsize=12)
ax.set_ylabel('Tên sản phẩm', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Xu hướng doanh thu theo tháng
monthly_revenue = (
    retail_clean.groupby('YearMonth')['Revenue']
    .sum()
    .reset_index()
)
monthly_revenue['YearMonth_str'] = monthly_revenue['YearMonth'].astype(str)

plt.figure(figsize=(14, 6))
ax = sns.lineplot(data=monthly_revenue, x='YearMonth_str', y='Revenue',
                  marker='o', linewidth=2.5, color='steelblue')
ax.fill_between(range(len(monthly_revenue)), monthly_revenue['Revenue'],
                alpha=0.15, color='steelblue')
ax.set_title('Monthly Revenue Trend - Online Retail', fontsize=15)
ax.set_xlabel('Tháng', fontsize=12)
ax.set_ylabel('Tổng doanh thu (GBP)', fontsize=12)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x/1e3:.0f}K'))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Số đơn hàng độc nhất theo tháng
monthly_orders = (
    retail_clean.groupby('YearMonth')['InvoiceNo']
    .nunique()
    .reset_index()
)
monthly_orders.columns = ['YearMonth', 'NumOrders']
monthly_orders['YearMonth_str'] = monthly_orders['YearMonth'].astype(str)

plt.figure(figsize=(14, 5))
ax = sns.barplot(data=monthly_orders, x='YearMonth_str', y='NumOrders', palette='Blues_d')
ax.set_title('Number of Orders per Month - Online Retail', fontsize=15)
ax.set_xlabel('Tháng', fontsize=12)
ax.set_ylabel('Số hóa đơn độc nhất', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Boxplot phân bố Quantity và UnitPrice (lọc outlier ở mức 99th percentile)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

q_filtered = retail_clean[retail_clean['Quantity'] <= retail_clean['Quantity'].quantile(0.99)]
p_filtered = retail_clean[retail_clean['UnitPrice'] <= retail_clean['UnitPrice'].quantile(0.99)]

sns.boxplot(data=q_filtered, x='Quantity', ax=axes[0], color='skyblue')
axes[0].set_title('Phân bố Quantity (giới hạn 99th percentile)', fontsize=11)
axes[0].set_xlabel('Số lượng mỗi giao dịch')

sns.boxplot(data=p_filtered, x='UnitPrice', ax=axes[1], color='salmon')
axes[1].set_title('Phân bố Unit Price (giới hạn 99th percentile)', fontsize=11)
axes[1].set_xlabel('Đơn giá (GBP)')

plt.suptitle('Boxplot - Quantity & UnitPrice (Online Retail)', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Heatmap doanh thu theo quốc gia và tháng (top 8 quốc gia)
top8_countries = (
    retail_clean.groupby('Country')['Revenue']
    .sum()
    .sort_values(ascending=False)
    .head(8)
    .index.tolist()
)

heatmap_df = (
    retail_clean[retail_clean['Country'].isin(top8_countries)]
    .groupby(['Country', 'YearMonth'])['Revenue']
    .sum()
    .reset_index()
)
heatmap_df['YearMonth'] = heatmap_df['YearMonth'].astype(str)
heatmap_pivot = heatmap_df.pivot(index='Country', columns='YearMonth', values='Revenue').fillna(0)

plt.figure(figsize=(16, 6))
sns.heatmap(heatmap_pivot / 1000, annot=True, fmt='.0f', cmap='YlOrRd',
            linewidths=0.5, cbar_kws={'label': 'Doanh thu (GBP nghìn)'})
plt.title('Heatmap Doanh thu theo Quốc gia & Tháng (Top 8) - đơn vị: GBP K', fontsize=13)
plt.xlabel('Tháng')
plt.ylabel('Quốc gia')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Pie chart tỷ lệ doanh thu theo quốc gia (top 5 + Others)
country_rev = retail_clean.groupby('Country')['Revenue'].sum().sort_values(ascending=False)
top5 = country_rev.head(5)
others = country_rev.iloc[5:].sum()
pie_data = pd.concat([top5, pd.Series({'Others': others})])

plt.figure(figsize=(9, 9))
plt.pie(pie_data.values,
        labels=pie_data.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=sns.color_palette('Set2', len(pie_data)))
plt.title('Tỷ lệ doanh thu theo quốc gia - Online Retail', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Quantity vs Revenue (top 5 quốc gia)
import numpy as np

top5_countries = (
    retail_clean.groupby('Country')['Revenue']
    .sum().sort_values(ascending=False).head(5).index.tolist()
)
scatter_df = retail_clean[retail_clean['Country'].isin(top5_countries)].copy()
# Lọc bỏ outlier để biểu đồ dễ đọc
scatter_df = scatter_df[scatter_df['Revenue'] <= scatter_df['Revenue'].quantile(0.99)]

plt.figure(figsize=(12, 6))
ax = sns.scatterplot(data=scatter_df, x='Quantity', y='Revenue',
                     hue='Country', alpha=0.4, s=20)
ax.set_title('Mối quan hệ Quantity và Revenue (Top 5 Quốc gia)', fontsize=15)
ax.set_xlabel('Số lượng', fontsize=12)
ax.set_ylabel('Doanh thu (GBP)', fontsize=12)
plt.tight_layout()
plt.show()
```

## 1.3. PHÂN TÍCH ĐƠN BIẾN VÀ HAI BIẾN

### 1.3.1. Ôn lý thuyết

- **Phân tích đơn biến**: phân tích 1 biến tại một thời điểm (histogram, boxplot, pie chart, countplot).
- **Phân tích hai biến**: khám phá mối quan hệ giữa 2 biến (scatter plot, crosstab, pivot table, heatmap).
- **Correlation vs Covariance**: 
  - Covariance: đo hướng quan hệ, phụ thuộc đơn vị.
  - Correlation: chuẩn hóa covariance, giá trị trong [-1, 1], không phụ thuộc đơn vị.

### 1.3.3. Bài tập thực hành 1 - SweetViz

```python
# Cài đặt: pip install sweetviz
import pandas as pd
import sweetviz as sv

marketing_data = pd.read_csv("data/marketing_campaign.csv", sep="\t")

# Tạo báo cáo EDA
report = sv.analyze(marketing_data)
report.show_html("Reports/sweetviz_report.html")
print("SweetViz report đã được lưu tại Reports/sweetviz_report.html")
```

```python
# So sánh hai nhóm: khách hàng có và không có trẻ em
marketing_data['HasKid'] = marketing_data['Kidhome'].apply(lambda x: 'Has Kid' if x > 0 else 'No Kid')

report_compare = sv.compare_intra(marketing_data,
                                   marketing_data['HasKid'] == 'Has Kid',
                                   ["Has Kid", "No Kid"])
report_compare.show_html("Reports/sweetviz_compare.html")
print("SweetViz compare report đã được lưu.")
```
